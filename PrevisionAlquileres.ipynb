# Simulación de precios de alquileres en Barcelona para el próximo año

El confinamiento y las malas expectativas económicas pueden afectar el precio de alquiler en Barcelona. Para poder cuantificar ese impacto se hace una previsión del precio del metro cuadrado, intentando explicarlo con variables económicas recogidas por la Oficina de Estadística del Ayuntamiento de Barcelona.

```python
import pandas as pd 
from fbprophet import Prophet
import matplotlib.pyplot as plt  # Matlab-style plotting
import seaborn as sns
import statsmodels.api as sm

%matplotlib inline
color = sns.color_palette()
sns.set_style('darkgrid')
```

Se importan los datos en un Dataframe que utilizaremos para realizar un análisis previo. Los datos son mensuales; el precio medio del alquiler se ha conseguido en la página de alquileres idealista.com (https://www.idealista.com/sala-de-prensa/informes-precio-vivienda/alquiler/cataluna/barcelona-provincia/barcelona/) y el resto en la web de la Oficina de Estadística del Ayuntamiento (https://www.bcn.cat/estadistica/catala/actual/index.htm).
Las posibles variables explicativas de los cambios en los precios de alquiler son:
1. El paro en la ciudad de Barcelona.
2. Número de vuelos llegados a Barcelona desde UE
3. Número de vuelos llegados a Barcelona desde el resto del mundo
4. Inflación en la provincia de Barcelona (2016 - 100)
5. Número de hipotecas firmadas
6. Importe medio de las hipotecas

El paro es un buen termometro de la economía ya que no se dispone del PIB mensual de un municipio. El número de aeronaves refleja la tendencia del turismo, la inflación indica la evolución de los precios y las variables relacionadas con las hipotecas muestran el estado del sector inmobiliario en particular. 

```python
filepath = "/home/ivan/rent_forecast/datos_bcn.csv"
df = pd.read_csv(filepath, sep=';')
df.head()
```

Observamos que el precio por metro cuadrado está en un momento de ciclo alcista. El paro comparte la tendencia del precio, pero al revés: sube hasta el 2014 y luego disminuye hasta el inicio de la cuarentena.
Parece que la llegada de los vuelos no será un buen predictor del precio ya que aumenta año tras año y no sigue una tendencia cíclica. La inflación, con base 100 = inflación en 2016, tiene una pendiente positiva por lo que también la descartamos.
El gráfico del número de hipotecas no se recupera de la caída que sufrió en 2014 tras el cambio de ley hipotecaria; el importe de hipotecas parece que sigue bastante a la par la tendencia del precio de alquiler.

Solo dos variables nos podrían servir para intentar preveer el precio de alquiler (a parte del histórico de datos de la misma): el paro y el importe de la hipoteca. Como están fuertemente correlacionadas, nos tenemos que quedar con una de ellas. Se escoge el paro ya que es una variable relacionada con la actividad económica en general y no solo con el mercado inmobiliario. Como finalmente no utilizamos la variable ImporteHipoteca no es necesario tratar la falta de valor de abril, el NaN que aparece en el paso anterior y que es el único que tenemos en nuestros datos.

Para intentar preveer la evolución de los precios de alquiler utilizaremos el algoritmo Prophet de Facebook (https://facebook.github.io/prophet/docs/quick_start.html#python-api). El modelo a simular será:
$$
\begin {align}
y(t) = g(t) + s(t) + h(t) + paro(t) + \epsilon (t)
\end{align}
$$
En el que $g(t)$ es la parte cíclica de la variable, la $s(t)$ es la tendencial, $h(t)$ refleja el impacto de las vacaciones y $\epsilon(t)$ es el error debido a todos los factores que no están incluidos en el modelo. Además, añadimos el regresor auxiliar $paro(t)$. Si se quiere saber más sobre el modelo, se puede consultar la publicación en el siguiente enlace: https://peerj.com/preprints/3190/

Prophet sigue el mismo procedimiento que la librería scikit-learn: se ajusta el modelo y luego se realizan las previsiones. No es necesario normalizar los datos, la librería lo hará por nosotros cuando ajuste el modelo. Cabe destacar que el prophet nos fuerza a: 
- Tener las fechas como datos, no como índice. Además tienen que estar en el formato datatime y el nombre  de columna tiene que se ds
- La variable a explicar, en nuestro caso el precio, tiene que llamarse y.
- El regresor puede tener cualquier nombre. 

```python
prophet_df = pd.DataFrame({"ds": df["Fecha"], "y": df["Precio"], "paro": df["Paro"]})
prophet_df
```

Se ajusta el modelo.

```python
m = Prophet()
m.add_regressor("paro")
m.fit(prophet_df)
```

Para realizar una previsión para los próximos 20 meses hay que tener:
- La variable ds con las fechas de los próximos meses
- La variable paro con una previsión de los parados para los próximos veinte meses

Se ha estimado una evolución del paro parecida a la recesión del 2012. Es posible que en un futuro se amplie el trabajo con mas escenarios. 

```python
future = m.make_future_dataframe(periods=20, freq="M")
esc_data = [92616, 94358, 94438, 92312,
96091, 99595, 101062, 102608, 105787, 107122,
110326,112309, 108345, 100037,95478, 98663,
99405,100200,103465,99030]
esc_index = range(142,162)
paro_esc= pd.Series(esc_data, index = esc_index)
future["paro"] = df["Paro"].append(paro_esc)
```

Se realiza la estimación. Prophet nos proporciona tanto los valores mínimos como los máximos. Se realiza la 
representación gráfica de los resultados. 

```python
forecast = m.predict(future)
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(12))
m.plot_components(forecast)
```

Observamos que la caída de los precios no será tan acusada como podríamos pensar en un momento. El precio previsto para el enero de 2021 sigue siendo un 2.4% mas elevado que el de 2020. Estos resultados se podrían explicar por:
- La entrada al mercado de los pisos que antes estaban en el mercado de alquiler turístico.
- La poca elasticidad del mercado inmobiliario
